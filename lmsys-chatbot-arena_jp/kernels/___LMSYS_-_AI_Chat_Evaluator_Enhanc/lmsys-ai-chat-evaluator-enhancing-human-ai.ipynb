```python
# å¿œç­”ã¨ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’é€£çµã—ã¦ãƒˆãƒ¼ã‚¯ãƒ³åŒ–ã™ã‚‹
texts = train_df['prompt'] + ' ' + train_df['response_a'] + ' ' + train_df['response_b']  # ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã¨å¿œç­”ã‚’çµåˆ
tokenizer = Tokenizer()  # ãƒˆãƒ¼ã‚¯ãƒŠã‚¤ã‚¶ãƒ¼ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã‚’ä½œæˆ
tokenizer.fit_on_texts(texts)  # ãƒ†ã‚­ã‚¹ãƒˆã«åŸºã¥ã„ã¦ãƒˆãƒ¼ã‚¯ãƒŠã‚¤ã‚¶ãƒ¼ã‚’ãƒ•ã‚£ãƒƒãƒˆã•ã›ã‚‹

# ãƒ†ã‚­ã‚¹ãƒˆã‚’ã‚·ãƒ¼ã‚±ãƒ³ã‚¹ã«å¤‰æ›ã™ã‚‹é–¢æ•°
def text_to_sequences(texts):
    sequences = tokenizer.texts_to_sequences(texts)  # ãƒ†ã‚­ã‚¹ãƒˆã‚’ã‚·ãƒ¼ã‚±ãƒ³ã‚¹ã«å¤‰æ›
    return pad_sequences(sequences, maxlen=500)  # maxlenã«åŸºã¥ã„ã¦ãƒ‘ãƒ‡ã‚£ãƒ³ã‚°å‡¦ç†ã‚’è¡Œã†

# è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã®å‡¦ç†
X_prompt_a = text_to_sequences(train_df['prompt'])  # ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’ã‚·ãƒ¼ã‚±ãƒ³ã‚¹ã«å¤‰æ›
X_response_a = text_to_sequences(train_df['response_a'])  # å¿œç­”Aã‚’ã‚·ãƒ¼ã‚±ãƒ³ã‚¹ã«å¤‰æ›
X_response_b = text_to_sequences(train_df['response_b'])  # å¿œç­”Bã‚’ã‚·ãƒ¼ã‚±ãƒ³ã‚¹ã«å¤‰æ›

# ãƒ©ãƒ™ãƒ«ã®å–å¾—
y_a = train_df['winner_model_a']  # ãƒ¢ãƒ‡ãƒ«Aã®å‹è€…ãƒ©ãƒ™ãƒ«
y_b = train_df['winner_model_b']  # ãƒ¢ãƒ‡ãƒ«Bã®å‹è€…ãƒ©ãƒ™ãƒ«
y_tie = train_df['winner_tie']  # å¼•ãåˆ†ã‘ãƒ©ãƒ™ãƒ«

# è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã®åˆ†å‰²
X_train_a, X_val_a, y_train_a, y_val_a = train_test_split(X_prompt_a, y_a, test_size=0.2, random_state=42)  # è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã¨è©•ä¾¡ãƒ‡ãƒ¼ã‚¿ã«åˆ†å‰²
X_train_b, X_val_b, y_train_b, y_val_b = train_test_split(X_prompt_a, y_b, test_size=0.2, random_state=42)  # è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã¨è©•ä¾¡ãƒ‡ãƒ¼ã‚¿ã«åˆ†å‰²
X_train_tie, X_val_tie, y_train_tie, y_val_tie = train_test_split(X_prompt_a, y_tie, test_size=0.2, random_state=42)  # è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã¨è©•ä¾¡ãƒ‡ãƒ¼ã‚¿ã«åˆ†å‰²
```

5. ãƒ¢ãƒ‡ãƒ«ã®æ§‹ç¯‰
</h1>


```python
# ãƒ¢ãƒ‡ãƒ«ã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã‚’å®šç¾©ã™ã‚‹
vocab_size = len(tokenizer.word_index) + 1  # ãƒœã‚­ãƒ£ãƒ–ãƒ©ãƒªãƒ¼ã®ã‚µã‚¤ã‚ºã‚’å–å¾—
embedding_dim = 100  # åŸ‹ã‚è¾¼ã¿æ¬¡å…ƒ
max_len = 500  # æœ€å¤§ã‚·ãƒ¼ã‚±ãƒ³ã‚¹é•·

# å…¥åŠ›ã®å®šç¾©
input_prompt = Input(shape=(max_len,))  # ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆå…¥åŠ›
input_response_a = Input(shape=(max_len,))  # å¿œç­”Aå…¥åŠ›
input_response_b = Input(shape=(max_len,))  # å¿œç­”Bå…¥åŠ›

# åŸ‹ã‚è¾¼ã¿å±¤ã®å®šç¾©
embedding_layer = Embedding(vocab_size, embedding_dim, input_length=max_len)

# å…¥åŠ›ã‚’åŸ‹ã‚è¾¼ã¿ãƒ™ã‚¯ãƒˆãƒ«ã«å¤‰æ›
embedded_prompt = embedding_layer(input_prompt)  # ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã®åŸ‹ã‚è¾¼ã¿
embedded_response_a = embedding_layer(input_response_a)  # å¿œç­”Aã®åŸ‹ã‚è¾¼ã¿
embedded_response_b = embedding_layer(input_response_b)  # å¿œç­”Bã®åŸ‹ã‚è¾¼ã¿

# LSTMå±¤ã®å®šç¾©
lstm_layer = LSTM(64)  # LSTMãƒ¦ãƒ‹ãƒƒãƒˆæ•°ã¯64

# å…¥åŠ›ã‚’LSTMå±¤ã«é€šã™
encoded_prompt = lstm_layer(embedded_prompt)  # ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã®ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰
encoded_response_a = lstm_layer(embedded_response_a)  # å¿œç­”Aã®ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰
encoded_response_b = lstm_layer(embedded_response_b)  # å¿œç­”Bã®ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰

# ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã¨ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰ã•ã‚ŒãŸå¿œç­”ã‚’é€£çµã™ã‚‹
concat_a = Concatenate()([encoded_prompt, encoded_response_a])  # å¿œç­”Aã¨ã®é€£çµ
concat_b = Concatenate()([encoded_prompt, encoded_response_b])  # å¿œç­”Bã¨ã®é€£çµ

# åˆ†é¡ç”¨ã®å…¨çµåˆå±¤
dense_layer = Dense(64, activation='relu')  # å…¨çµåˆå±¤ï¼ˆReLUæ´»æ€§åŒ–é–¢æ•°ï¼‰

# å‡ºåŠ›å±¤ã®å®šç¾©
output_a = Dense(1, activation='sigmoid')(dense_layer(concat_a))  # å¿œç­”Aã®å‹è€…ã‚’äºˆæ¸¬
output_b = Dense(1, activation='sigmoid')(dense_layer(concat_b))  # å¿œç­”Bã®å‹è€…ã‚’äºˆæ¸¬
output_tie = Dense(1, activation='sigmoid')(dense_layer(concat_a))  # å¼•ãåˆ†ã‘ã®å‹è€…ã‚’äºˆæ¸¬

# ãƒ¢ãƒ‡ãƒ«ã®å®šç¾©
model = Model(inputs=[input_prompt, input_response_a, input_response_b], 
              outputs=[output_a, output_b, output_tie])  # ãƒ¢ãƒ‡ãƒ«ã®å…¥åŠ›ã¨å‡ºåŠ›ã‚’è¨­å®š

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])  # ãƒ¢ãƒ‡ãƒ«ã‚’ã‚³ãƒ³ãƒ‘ã‚¤ãƒ«

# ãƒ¢ãƒ‡ãƒ«ã®è¦ç´„ï¼ˆæ§‹é€ ï¼‰ã‚’è¡¨ç¤º
model.summary()
```

```python
# ãƒ¢ãƒ‡ãƒ«ã®ã‚³ãƒ³ãƒ‘ã‚¤ãƒ«
model.compile(
    optimizer='adam',  # æœ€é©åŒ–æ‰‹æ³•ã«Adamã‚’ä½¿ç”¨
    loss=['categorical_crossentropy', 'categorical_crossentropy', 'categorical_crossentropy'],  # æå¤±é–¢æ•°ã«ã‚«ãƒ†ã‚´ãƒªã‚«ãƒ«ã‚¯ãƒ­ã‚¹ã‚¨ãƒ³ãƒˆãƒ­ãƒ”ãƒ¼ã‚’ä½¿ç”¨
    metrics=[['accuracy'], ['accuracy'], ['accuracy']]  # å„å‡ºåŠ›ã®ç²¾åº¦ã‚’è©•ä¾¡æŒ‡æ¨™ã§æŒ‡å®š
)

# ãƒ¢ãƒ‡ãƒ«ã®ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°
history = model.fit(
    [X_train_a, X_train_b, X_train_b],  # è¨“ç·´ãƒ‡ãƒ¼ã‚¿
    [y_train_a, y_train_b, y_train_tie],  # è¨“ç·´ãƒ©ãƒ™ãƒ«
    validation_data=([X_val_a, X_val_b, X_val_b], [y_val_a, y_val_b, y_val_tie]),  # æ¤œè¨¼ãƒ‡ãƒ¼ã‚¿
    epochs=5,  # ã‚¨ãƒãƒƒã‚¯æ•°ã¯å¿…è¦ã«å¿œã˜ã¦èª¿æ•´
    batch_size=32  # ãƒãƒƒãƒã‚µã‚¤ã‚º
)
```
